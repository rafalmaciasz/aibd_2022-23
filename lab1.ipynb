Import niezbędnych modułów

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Definicja funkcji i wywłanie jej na trzech wektorach

```python
x = [np.arange(-1, 1), np.arange(-6, 6), np.arange(0, 5)]
fx = lambda x: x ** 2 + 5

for i in range(3):
    plt.plot(x[i], fx(x[i]))
    plt.title("Wykres nr{}".format(i + 1))
    plt.legend(["f(x)"])
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()
```
