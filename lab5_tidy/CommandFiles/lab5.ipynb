## Create a function to compute the datetime when each listing was published

```python
def reDate(date: str, week: int) -> str:
    date = date.rstrip()
    year, month, day = int(date[:4]), int(date[5:7]), int(date[8:])
    date = datetime(year, month, day)
    for i in range(week):
        try:
            date = datetime(date.year, date.month, date.day + 7)
        except:
            if date.month in [1, 3, 5, 7, 8, 10]:
                # 31-days months
                date = datetime(date.year, date.month + 1, date.day + 7 - 31)
            elif date.month in [4, 6, 9, 11]:
                # 30-days months
                date = datetime(date.year, date.month + 1, date.day + 7 - 30)
            elif date.month == 2:
                # February
                date = datetime(date.year, date.month + 1, date.day + 7 - 28)
            else:
                # December
                date = datetime(date.year + 1, 1, date.day + 7 - 31)
    return "{:4}-{:2}-{:2}".format(date.year, date.month, date.day).replace(" ", "0")
```

## Obtain a data about a place in ranking

```python
dfWeeks = df[df.columns[7:]]
dfWeeks
```

## Create a dataframe with ranking information

```python
rankColumns = ["artist", "track", "date", "week", "rank"]
dfRank = pd.DataFrame(columns=rankColumns)

for i in range(len(dfWeeks.index)):
    for j in range(len(dfWeeks.values[i])):
        if type(dfWeeks[dfWeeks.columns[j]][i]) == str:
            dfWeeks[dfWeeks.columns[j]][i] = int(dfWeeks[dfWeeks.columns[j]][i])
        if dfWeeks[dfWeeks.columns[j]][i] > 0:
            toAppend = pd.DataFrame({
                "artist": df["artist.inverted"][i], 
                "track": df["track"][i],
                "date": reDate(df["date.entered"][i], j + 1), 
                "week": j + 1, 
                "rank": int(dfWeeks[dfWeeks.columns[j]][i])}, index=[0])
            dfRank = pd.concat([dfRank, toAppend], ignore_index=True)
            
dfRank.sort_values(["artist", "track"]).reset_index().drop(labels=["index"], axis=1)
```

## Save the dataframe to a .csv file

```python
dfRank.to_csv("../AnalysisData/billboardRank.csv", index=False)
```
